```python
input_file_path_1="file:///C:/Users/ckp43_000/Documents/generic-food.csv"
file_rdd_1=sc.textFile(input_file_path_1)
input_file_path_2="file:///C:/Users/ckp43_000/Documents/flavors.csv"
file_rdd_2=sc.textFile(input_file_path_2)
filter_file_rdd_1=file_rdd_1.filter(lambda row:row!="FOOD NAME,SCIENTIFIC NAME,GROUP,SUB GROUP")
filter_file_rdd_2=file_rdd_2.filter(lambda row:row !="GROUP,SUB GROUP")
```

```python
split_food_rdd=filter_file_rdd_1.map(lambda row:(row.split(',')[2],row.split(',')[3]))
split_food_rdd.collect()
```

```python
split_flavors_rdd=filter_file_rdd_2.map(lambda row:(row.split(',')[0],row.split(',')[1]))
split_flavors_rdd.collect()
```

```python

reduce_food_rdd=filter_file_rdd_1\
    .flatMap(lambda e:e.split(',')) \
    .map(lambda e: (e, 1))\
    .reduceByKey(lambda x,y:x+y)
print(" \n reduced food rdd :")
for ele in reduce_food_rdd.collect():
    print(ele)

```

```python
reduce_flavors_rdd=filter_file_rdd_2\
    .flatMap(lambda e:e.split(','))\
    .map(lambda e:(e,1))\
    .reduceByKey(lambda x,y:x+y)
print(" \n reduce flavors file rdd :")
for ele in reduce_flavors_rdd.collect():
    print(ele)
```
